```python
import pandas as pd
import json
import os
```

```python
results_dir = 'results/'
results_dict = {}
for filename in [a for a in os.listdir(results_dir) if a.endswith('.json')]:
    with open(results_dir+filename) as file:
        data = json.load(file)
        results_dict[filename[:-5]] = data
results_dict.keys()
```

```python
dfEUI = pd.DataFrame(results_dict['EntradaUi']).T.sort_values(['mae', 'mape'])
dfEUTI = pd.DataFrame(results_dict['EntradaUti']).T.sort_values(['mae', 'mape'])
dfSUI = pd.DataFrame(results_dict['SaidaUi']).T.sort_values(['mae', 'mape'])
dfSUTI = pd.DataFrame(results_dict['SaidaUti']).T.sort_values(['mae', 'mape'])
```
